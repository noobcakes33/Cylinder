```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import mpl_toolkits.mplot3d.art3d as art3d
import pandas as pd

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

df = pd.read_csv('cylinder.csv')

for i in range(int(df['min_cylinder_height']), int(df['max_cylinder_height'])):
    p = Circle((df['centre_x'], df['centre_y']), df['radius'], color= 'b', fill=0)
    ax.add_patch(p)
    art3d.pathpatch_2d_to_3d(p, z=i , zdir="z")


ax.set_xlim(int(df['centre_x']) - int(df['radius']) , int(df['centre_x']) + int(df['radius']))
ax.set_ylim(int(df['centre_y']) - int(df['radius']), int(df['centre_y']) + int(df['radius']))
ax.set_zlim(int(df['min_cylinder_height']), int(df['max_cylinder_height']))
plt.show()   
fig.savefig('cylinderPlot.png')
```
